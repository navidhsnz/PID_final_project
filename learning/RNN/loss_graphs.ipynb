```python
import matplotlib.pyplot as plt
```

```python
def plot_losses(train_loss_file, val_loss_file, y_range, title):
    with open(train_loss_file, "r") as file:
        train_losses = [float(line.strip()) for line in file.readlines()]
    with open(val_loss_file, "r") as file:
        val_losses = [float(line.strip()) for line in file.readlines()]

    plt.figure(figsize=(6, 4))
    plt.plot(train_losses, label="Training Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.ylim(y_range[0], y_range[1]) 
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# RNN1

plot_losses("RNN1/train_losses_RNN1.txt", "RNN1/val_losses_RNN1.txt", y_range = (0,0.006), title="RNN1: Training and Validation Loss Over Epochs")
```

```python
# RNN2

plot_losses("RNN2/train_losses_RNN2.txt", "RNN2/val_losses_RNN2.txt", y_range = (0,0.006), title="RNN2: Training and Validation Loss Over Epochs")
```
