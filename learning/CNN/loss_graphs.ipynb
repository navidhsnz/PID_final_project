```python
import matplotlib.pyplot as plt
```

```python
def plot_losses(train_loss_file, val_loss_file, y_range, title):
    with open(train_loss_file, "r") as file:
        train_losses = [float(line.strip()) for line in file.readlines()]
    with open(val_loss_file, "r") as file:
        val_losses = [float(line.strip()) for line in file.readlines()]

    plt.figure(figsize=(6, 4))
    plt.plot(train_losses, label="Training Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.ylim(y_range[0], y_range[1]) 
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# CNN1

plot_losses("CNN1/train_losses_CNN1.txt", "CNN1/val_losses_CNN1.txt", y_range = (0,0.002), title="CNN1: Training and Validation Loss Over Epochs")
```

```python
# CNN2

plot_losses("CNN2/train_losses_CNN2.txt", "CNN2/val_losses_CNN2.txt", y_range = (0,0.002), title="CNN2: Training and Validation Loss Over Epochs")
```
